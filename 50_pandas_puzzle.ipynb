Apply a custom function row-wise

```python
def Total_stat(Total):
    if Total < 100:
        return 'low'
    elif Total < 500:
        return 'mid'
    else:
        return 'high'

df['Total_stat'] = df['Total'].apply(Total_stat)
df
```
